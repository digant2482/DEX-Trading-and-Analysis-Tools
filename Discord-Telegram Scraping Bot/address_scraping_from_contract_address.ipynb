```python
import requests
import pandas as pd
import json
import datetime
import time

unique_address_dictionary_list = {'contract_address': ['unique address for a contract']}
unique_address_dictionary = {'address': 'occurrences'}
address_list = list(set(pd.read_excel('/content/data_2022-12-01 to 21.xlsx')['To'].tolist()))
```

```python
def get_unique_addresses_dictionary(unique_address_list):
    global unique_address_dictionary
    for address in unique_address_list:
        try:
            unique_address_dictionary[address] = unique_address_dictionary[address] + 1
        except:
            unique_address_dictionary[address] = 1
```

```python
flag = 0
count = 0 

for address in address_list:
    addresses_for_the_contract = []
    if address == '0xb7e632272a9a01e0d8d71ecb340408c4bb2dc530'.lower():
        flag = 1
        continue

    if flag == 1:
        count += 1 
        print('contract_address :' + address)
        time_start = datetime.datetime.now().timestamp()
        url = "https://api.etherscan.io/api?module=account&action=txlist&address={}&startblock=0&endblock=99999999&page=1&offset=10000&sort=asc&apikey=TRT149Y52NFS2DYR6HPJF3QW2P3DBTTAJ2".format(address)
        response = requests.request("GET", url)
        json_data = json.loads(response.text)['result']
    
        if len(json_data) > 19:
            for i in json_data[:200]:
                if i['functionName'] == 'approve(address _spender, uint256 _value)' and i['from'] not in addresses_for_the_contract and i['from'] != address:
                        addresses_for_the_contract.append(i['from'])
            unique_address_dictionary_list[address] = addresses_for_the_contract
            get_unique_addresses_dictionary(addresses_for_the_contract)
        time_end = datetime.datetime.now().timestamp()

        if (time_end - time_start) < 1:
            time.sleep(1 - (time_end - time_start))
```
