```python
from telethon.sync import TelegramClient
import datetime
import pandas as pd

api_id = '15585886'
api_hash = 'e8ce9f868a7c68e4488e3563a7de601f'

chats = ['DEXTNewPairsBot']

client =  TelegramClient(None, api_id, api_hash)

df = pd.DataFrame()

for chat in chats:
    async with TelegramClient(None, api_id, api_hash) as client:
      # Date to be changed
        async for message in client.iter_messages(chat, offset_date=datetime.datetime(2022, 12, 1) , reverse=True):
            data = { "group" : chat, "sender" : message.sender_id, "text" : message.text, "date" : message.date}

            temp_df = pd.DataFrame(data, index=[1])
            df = df.append(temp_df)

contract_list = [messageString.split("contract:\n")[1].split("\n")[0].lower() for messageString in df['text'].tolist()]
new_df = pd.DataFrame(contract_list)

df['date'] = df['date'].dt.tz_localize(None).tolist()
new_df['release_date'] = df['date'].tolist()

contract_date_list = [str(x).split()[0] for x in df['date'].tolist()]
# new_df.to_excel("data_{}.xlsx".format(datetime.datetime(2022, 12, 16)), index=False)
```

```python
import requests

# Date to be changed
DATE = '2022-12-01'
CHANNELID = '727212168138981526'
HEADERS = {'authorization': 'ODY1OTIyNTY3NzkzOTM0MzQ3.GwgCMm.Iwv1Z4er7WcBC8CBKhFOGyMOM4BQ3tllmg4QaQ'}
LIMIT=100
all_messages = []

def add_address(json_list):
  for message in json_list:
    new_address = message['content'].split('Token: <')[1].split('>')[0].split('/')[-1].lower()
    new_date = message['timestamp'].split('T')[0]
    if new_address not in contract_list and new_date >= DATE:
      contract_list.append(new_address)
      contract_date_list.append(new_date)
    elif new_date < DATE:
      return False
  return True

r = requests.get(f'https://discord.com/api/v9/channels/{CHANNELID}/messages?limit={LIMIT}',headers=HEADERS)
all_messages.extend(r.json())
is_added = add_address(r.json())

while len(r.json()) == LIMIT and is_added:
    last_message_id = r.json()[-1].get('id')
    r = requests.get(f'https://discord.com/api/v9/channels/{CHANNELID}/messages?limit={LIMIT}&before={last_message_id}',headers=HEADERS)
    all_messages.extend(r.json())
    is_added = add_address(r.json())
```

```python
new_df = pd.DataFrame(contract_list)
new_df['release_date'] = contract_date_list
# Date to be changed
new_df.to_excel("data_{}.xlsx".format(datetime.datetime(2022, 12, 1)), index=False)
```
